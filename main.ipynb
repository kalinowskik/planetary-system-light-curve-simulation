```python
import celestial_objects as co
import math as m    
from lightcurve import *
```

```python
def ComputePeriod(a1, a2, T2): #compute period according to Kepler's 3rd law
    T1=((a1/a2)**3*T2**2)**(1/2)
    return T1


p2p=ComputePeriod(0.04,0.025,4.91)
p3p=ComputePeriod(0.06,0.025,4.91)
```

## Define star and planets

```python
Sun=co.star(0.1,2900, 0.0007)
planet1=co.planet(albedo=0.35, radius=1, distance=0.025, beta=0.5, i=m.pi*0.98, Star=Sun, period=4.91, name='Lusitania')
planet2=co.planet(0.52, 1.2, 0.04, 0.3, m.pi*1.04, Sun, period=p2p, name='Megas')
planet3=co.planet(0.2, 0.9, 0.06, 0.4, m.pi*1.01, Sun, period=p3p, name='Ganges')
```

## Define observation of a system with just 1 planet

```python
first_star_system_observation=LC(PlanetList=[planet1],Star=Sun,SAMPLE_RATE=1/24/60*30,ObservationPeriod=365/2,name='first')
```

## Generate data for the first observation and do Fourier Transform

```python
first_star_system_observation.Signal(Clouds=0, sigma=0, filter='bessel_k', solar_day_duration=1, observing_night_duration=1)
first_star_system_observation.PlotComponents()
first_star_system_observation.FourierTransform(xupperlim=0.3, yupperlim=4e20)
```

# Multiple planet system

## Define observation for system with 3 planets

```python
second_star_system_observation=LC(PlanetList=[planet1,planet2,planet3],Star=Sun,SAMPLE_RATE=1/24/60*30,ObservationPeriod=365/2,name='second')
```

## Generate data for the multi planet system observation and do Fourier Transform

```python
second_star_system_observation.Signal(Clouds=0, sigma=0, filter='bessel_k', solar_day_duration=1, observing_night_duration=1)
second_star_system_observation.PlotComponents()
second_star_system_observation.FourierTransform(xupperlim=0.3, yupperlim=4e20)
```

# Add noise to the data

```python
second_star_system_observation.Signal(Clouds=0, sigma=1e17, filter='bessel_k', solar_day_duration=1, observing_night_duration=1)
second_star_system_observation.PlotComponents()
second_star_system_observation.FourierTransform(xupperlim=0.3, yupperlim=4e20)
```

## Make 20% of days cloudy so that there are gaps in the light curve

```python
second_star_system_observation.Signal(Clouds=0.2, sigma=0, filter='bessel_k', solar_day_duration=1, observing_night_duration=1)
second_star_system_observation.PlotComponents()
second_star_system_observation.FourierTransform(xupperlim=0.3, yupperlim=4e20)
```
